Teste sequencial em testes A/B refere-se a um método em que o teste é analisado em múltiplos pontos durante o experimento, em vez de esperar que todo o tamanho da amostra seja coletado e analisado no final. O objetivo do teste sequencial é avaliar os resultados mais cedo no processo para determinar se uma variante é significativamente melhor que a outra antes que o teste atinja sua duração ou tamanho de amostra planejado.

Pontos principais sobre o teste sequencial em A/B:

    Análise Intermediária: O teste é avaliado em diferentes estágios conforme os dados são coletados. Em vez de esperar que todos os participantes completem o experimento, a análise ocorre em estágios.

    Parada Antecipada: Se uma análise intermediária mostrar um vencedor claro (significância estatística), o teste pode ser interrompido mais cedo para economizar tempo e recursos. Se uma variante já mostrar resultados fortes, talvez não seja necessário continuar coletando dados.

    Ajustes para Múltiplas Análises: Testes estatísticos tradicionais assumem que você olha para os dados uma vez, depois que todos os dados são coletados. No entanto, no teste sequencial, ajustes são necessários para levar em conta o fato de que você está analisando os dados em vários pontos. Sem esses ajustes, o risco de encontrar um falso positivo (erro do Tipo I) aumenta a cada análise adicional.

    Métodos Estatísticos: Técnicas estatísticas especiais, como testes de razão de probabilidade sequenciais (SPRT) ou métodos bayesianos, são frequentemente usadas para controlar o aumento do risco de falsos positivos ao realizar uma análise sequencial.

Vantagens:

    Resultados mais rápidos: Você pode chegar a conclusões mais rapidamente do que com métodos tradicionais de testes A/B.
    Eficiência de recursos: O teste pode ser interrompido mais cedo se houver evidências claras de um vencedor, economizando tempo e recursos.

Desvantagens:

    Complexidade: Requer técnicas estatísticas mais avançadas para ajustar corretamente as análises múltiplas.
    Risco de interpretação incorreta: Se não for feito de maneira cuidadosa, o teste sequencial pode levar a conclusões erradas, especialmente se os critérios de parada não estiverem bem definidos.

Exemplo:

Imagine que você está realizando um teste A/B para ver qual de dois designs de site tem melhor taxa de conversão. Em vez de esperar por 1.000 usuários em cada grupo, você pode olhar os resultados após 100 usuários, depois 200, e assim por diante. Se após 200 usuários você já descobrir que um design é estatisticamente melhor, pode interromper o teste antecipadamente e implementar o design vencedor.
Considerações importantes:

    Potência Estatística: Testes sequenciais podem afetar a potência estatística do seu teste. Técnicas especiais ou ajustes, como correções de Bonferroni ou funções de alocação de erro (Alpha Spending Functions), podem ser usadas para levar em consideração o fato de que você está analisando os dados várias vezes.
    Critérios de Parada: Defina as regras para quando interromper o teste antecipadamente (por exemplo, se um resultado atingir um determinado nível de significância estatística).

Em resumo, o teste sequencial permite que você analise os dados de um teste A/B em vários pontos, oferecendo a possibilidade de decisões mais rápidas, mas requer um planejamento cuidadoso e métodos estatísticos adequados para evitar conclusões enganosas.

O problema do "peeking" (ou "olhar antes da hora") em testes A/B refere-se à prática de olhar repetidamente para os resultados de um experimento antes que ele esteja completamente concluído. Esse comportamento pode levar a erros na interpretação dos resultados e aumentar a probabilidade de encontrar um falso positivo, ou seja, concluir que existe uma diferença significativa entre as variáveis quando, na realidade, isso é apenas uma flutuação aleatória.
Como o "peeking" acontece:

    Análises intermediárias: Em um experimento A/B, é comum querer saber os resultados logo após uma parte da amostra ser coletada. Isso pode levar a decisões prematuras sobre qual variante (A ou B) parece ser a melhor, sem esperar até o fim do experimento.

    Múltiplas verificações: Cada vez que você olha para os resultados e faz uma análise, você aumenta o risco de tirar conclusões erradas, especialmente se você fizer várias verificações ao longo do teste. Quanto mais vezes você "espiar" os resultados antes de terminar o experimento, maior é a chance de identificar algo que parece significativo apenas por acaso.

Conclusão:

O problema do "peeking" ocorre quando os resultados de um experimento A/B são analisados repetidamente durante a coleta de dados, o que pode levar a conclusões erradas e falsos positivos. Para evitar esse problema, é importante planejar adequadamente o experimento, usar métodos de correção estatística e evitar olhar para os resultados antes do momento apropriado.

A relação entre o problema do "peeking" (ou olhar repetido para os resultados) e os valores t e valores p é crucial para entender como o processo de analisar os resultados várias vezes pode distorcer as conclusões estatísticas.
Como o "peeking" afeta o valor t e o valor p:

    Valor t e p-valor no contexto de um teste A/B:
        O valor t mede a diferença entre as médias dos dois grupos em relação à variação dentro dos grupos. Quando você realiza um teste estatístico (como um teste t de Student), o valor t é usado para calcular o p-valor, que indica a probabilidade de observar uma diferença tão grande quanto a observada, assumindo que a hipótese nula (de que não há diferença entre os grupos) seja verdadeira.
        Se o p-valor for baixo (geralmente menor que 0,05), significa que a diferença observada entre os grupos é estatisticamente significativa, ou seja, é improvável que tenha ocorrido por acaso.

    O impacto do "peeking":
        Quando você realiza várias análises durante o experimento (ou seja, faz "peeking" nos resultados), o p-valor pode ser distorcido. Cada vez que você calcula o p-valor ao olhar para os dados, há uma chance de encontrar um valor p pequeno simplesmente devido a flutuações aleatórias nos dados. Isso pode levar a um falso positivo, onde você acredita que encontrou uma diferença significativa, quando na verdade, isso aconteceu por acaso.

    Aumento do risco de falso positivo (erro Tipo I):
        Cada vez que você verifica os dados, você aumenta a chance de cometer um erro do Tipo I (rejeitar a hipótese nula quando ela é verdadeira). O valor p foi originalmente calculado sob a suposição de que você só faria uma análise dos dados no final do experimento. Mas, ao olhar várias vezes para os resultados, a probabilidade de encontrar um valor p pequeno por acaso aumenta.
        Em termos simples, o valor p é uma probabilidade condicional: ele representa a probabilidade de observar os dados que você obteve (ou algo mais extremo), dado que a hipótese nula seja verdadeira. Se você olhar para os dados muitas vezes, a chance de encontrar um valor p significativo aumenta, mesmo que não haja uma verdadeira diferença entre os grupos.

    Ajuste para múltiplas verificações:
        Para corrigir esse problema, técnicas como correção de Bonferroni ou métodos de teste sequenciais são usadas. Essas técnicas ajustam o nível de significância (o critério para rejeitar a hipótese nula) para compensar as múltiplas análises feitas ao longo do tempo. Isso ajuda a reduzir a chance de encontrar um falso positivo.
        Exemplo: Se você olha para os dados cinco vezes durante o experimento, a probabilidade de cometer um erro Tipo I (falso positivo) se torna maior, então, para cada uma dessas verificações, o valor p necessário para rejeitar a hipótese nula precisaria ser mais rigoroso para manter o controle sobre a taxa de erro.

Resumo da relação entre "peeking", valor t e p-valor:

    Peeking aumenta o risco de falsos positivos ao permitir que os dados sejam analisados múltiplas vezes, levando a p-valores mais baixos do que os que ocorreriam se você esperasse até o final do experimento para analisá-los.
    O valor t e o p-valor são fundamentais para determinar se uma diferença observada entre grupos é estatisticamente significativa. Porém, ao olhar repetidamente para os resultados durante o experimento (peeking), você distorce a interpretação desses valores, o que pode levar a conclusões erradas.
    Ajustes estatísticos são necessários para corrigir esses erros, como testes sequenciais e correções para múltiplas verificações, que ajustam o nível de significância para compensar o "peeking".

Em resumo, o problema do "peeking" está diretamente relacionado ao risco de encontrar p-valores enganadores, especialmente quando há múltiplas verificações, o que pode levar a conclusões erradas sobre a significância estatística dos resultados.
